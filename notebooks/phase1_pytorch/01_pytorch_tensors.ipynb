## What is a Tensor?

A **tensor** is just a multi-dimensional array. That's it.

```
Scalar (0D tensor):  5
Vector (1D tensor):  [1, 2, 3]
Matrix (2D tensor):  [[1, 2], [3, 4]]
3D tensor:           A cube of numbers (like a CT volume)
4D tensor:           Batch of images (batch, channels, height, width)
```

PyTorch tensors are like NumPy arrays, but they can run on GPU and track gradients for training.

````python
import torch
import numpy as np

# Creating tensors
# ----------------

# From Python list
t1 = torch.tensor([1, 2, 3])
print(f"From list: {t1}")

# From NumPy
np_array = np.array([4, 5, 6])
t2 = torch.from_numpy(np_array)
print(f"From NumPy: {t2}")

# Zeros, ones, random
t3 = torch.zeros(3, 4)      # 3 rows, 4 columns
t4 = torch.ones(2, 3)
t5 = torch.randn(2, 3)      # Random from normal distribution

print(f"\nZeros (3x4):\n{t3}")
print(f"\nRandom (2x3):\n{t5}")
````

## Shape, Dtype, Device

Every tensor has three key properties:

| Property | What it means | Example |
|----------|--------------|--------|
| **shape** | Dimensions of the tensor | `(32, 3, 224, 224)` |
| **dtype** | Data type of elements | `float32`, `int64` |
| **device** | Where tensor lives | `cpu` or `cuda:0` |

````python
# Checking tensor properties
t = torch.randn(2, 3, 4)  # 2x3x4 tensor

print(f"Shape: {t.shape}")      # torch.Size([2, 3, 4])
print(f"Dtype: {t.dtype}")      # torch.float32 (default)
print(f"Device: {t.device}")    # cpu (default)
print(f"Number of dimensions: {t.ndim}")
print(f"Total elements: {t.numel()}")
````

````python
# Creating with specific dtype
t_float = torch.tensor([1, 2, 3], dtype=torch.float32)
t_int = torch.tensor([1, 2, 3], dtype=torch.int64)
t_half = torch.tensor([1, 2, 3], dtype=torch.float16)  # Half precision

print(f"float32: {t_float.dtype}")
print(f"int64: {t_int.dtype}")
print(f"float16: {t_half.dtype}")

# Converting dtype
t_converted = t_int.float()  # int64 -> float32
print(f"\nConverted: {t_converted.dtype}")
````

## Tensor Operations

Operations work element-wise by default:

````python
a = torch.tensor([1.0, 2.0, 3.0])
b = torch.tensor([4.0, 5.0, 6.0])

# Basic math (element-wise)
print(f"a + b = {a + b}")       # [5, 7, 9]
print(f"a * b = {a * b}")       # [4, 10, 18]
print(f"a ** 2 = {a ** 2}")     # [1, 4, 9]

# Useful functions
print(f"\nsum: {a.sum()}")      # 6
print(f"mean: {a.mean()}")      # 2
print(f"max: {a.max()}")        # 3
print(f"argmax: {a.argmax()}")  # 2 (index of max)
````

````python
# Matrix multiplication (the core of neural networks!)
# Two ways: @ operator or torch.matmul()

A = torch.randn(2, 3)  # 2x3 matrix
B = torch.randn(3, 4)  # 3x4 matrix

C = A @ B              # Result: 2x4 matrix
# Same as: C = torch.matmul(A, B)

print(f"A shape: {A.shape}")
print(f"B shape: {B.shape}")
print(f"A @ B shape: {C.shape}")

# Rule: (m x n) @ (n x p) = (m x p)
# The inner dimensions must match!
````

## Reshaping Tensors

You'll reshape tensors constantly. Key operations:

| Operation | What it does |
|-----------|-------------|
| `view()` | Reshape (must be contiguous) |
| `reshape()` | Reshape (handles non-contiguous) |
| `squeeze()` | Remove dimensions of size 1 |
| `unsqueeze()` | Add dimension of size 1 |
| `permute()` | Reorder dimensions |

````python
# Reshape examples
t = torch.arange(12)  # [0, 1, 2, ..., 11]
print(f"Original: {t.shape}")

# Reshape to 3x4
t_3x4 = t.view(3, 4)
print(f"\nReshaped to 3x4:\n{t_3x4}")

# Reshape to 2x2x3
t_2x2x3 = t.view(2, 2, 3)
print(f"\nReshaped to 2x2x3:\n{t_2x2x3}")

# Use -1 to infer dimension
t_auto = t.view(4, -1)  # 4 rows, PyTorch figures out columns
print(f"\nview(4, -1): {t_auto.shape}")
````

````python
# squeeze and unsqueeze - very common!

t = torch.randn(1, 3, 1, 4)
print(f"Original shape: {t.shape}")  # [1, 3, 1, 4]

# Remove all size-1 dimensions
t_squeezed = t.squeeze()
print(f"After squeeze(): {t_squeezed.shape}")  # [3, 4]

# Add a dimension at position 0
t2 = torch.randn(3, 4)
t2_with_batch = t2.unsqueeze(0)  # Add batch dimension
print(f"\nOriginal: {t2.shape}")
print(f"After unsqueeze(0): {t2_with_batch.shape}")  # [1, 3, 4]
````

## Indexing and Slicing

Same as NumPy:

````python
t = torch.arange(20).view(4, 5)
print(f"Tensor:\n{t}\n")

print(f"First row: {t[0]}")
print(f"Last row: {t[-1]}")
print(f"Element [1,2]: {t[1, 2]}")
print(f"First 2 rows: \n{t[:2]}")
print(f"Last 2 columns: \n{t[:, -2:]}")
````

## Moving to GPU

Moving tensors between CPU and GPU:

````python
# Check GPU availability
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# Create on CPU, move to GPU
t_cpu = torch.randn(1000, 1000)
t_gpu = t_cpu.to(device)

print(f"CPU tensor device: {t_cpu.device}")
print(f"GPU tensor device: {t_gpu.device}")

# Create directly on device
t_direct = torch.randn(1000, 1000, device=device)
print(f"Direct on device: {t_direct.device}")
````

## PyTorch vs NumPy

| PyTorch | NumPy | Notes |
|---------|-------|-------|
| `torch.tensor()` | `np.array()` | Create array |
| `.shape` | `.shape` | Same |
| `.view()` | `.reshape()` | Reshape |
| `@` or `.matmul()` | `@` or `.dot()` | Matrix multiply |
| `.to('cuda')` | N/A | GPU support |
| `.requires_grad` | N/A | Gradient tracking |

````python
# Converting between PyTorch and NumPy

# NumPy -> PyTorch
np_arr = np.array([1, 2, 3])
torch_tensor = torch.from_numpy(np_arr)

# PyTorch -> NumPy (must be on CPU!)
back_to_numpy = torch_tensor.numpy()

print(f"NumPy: {np_arr}")
print(f"PyTorch: {torch_tensor}")
print(f"Back to NumPy: {back_to_numpy}")

# Warning: they share memory!
np_arr[0] = 999
print(f"\nAfter modifying NumPy: {torch_tensor}")  # Also changed!
````

## Common Shapes in Deep Learning

| Data Type | Shape | Example |
|-----------|-------|--------|
| Batch of images | `(B, C, H, W)` | `(32, 3, 224, 224)` |
| Batch of 3D volumes | `(B, C, D, H, W)` | `(4, 1, 96, 96, 96)` |
| Batch of vectors | `(B, features)` | `(32, 512)` |
| Sequence data | `(B, seq_len, features)` | `(32, 100, 768)` |

**B** = batch size, **C** = channels, **H** = height, **W** = width, **D** = depth

````python
# Simulating your production input

# CT scan volume (like in your segmentation code)
ct_volume = torch.randn(1, 1, 96, 96, 96)  # batch=1, channels=1, 96x96x96
print(f"CT volume shape: {ct_volume.shape}")
print(f"  Batch size: {ct_volume.shape[0]}")
print(f"  Channels: {ct_volume.shape[1]}")
print(f"  Depth x Height x Width: {ct_volume.shape[2:]}")
````
